```python
def point(x,y,dashed = False, color = 'C1', marker = 'o', markersize = 18, annotate = '', fontsize = 14):
    
    ax = plt.gca()
    ax.plot([x],[y], marker = marker, markersize = markersize, color = color)
    ax.text(x,y, annotate, fontsize = fontsize)
    
    if dashed:
        ax.plot([0,x], [y,y], linestyle = 'dashed', color = 'gray')
        ax.plot([x,x], [0,y], linestyle = 'dashed', color = 'gray')
```

```python
def plot_cleaner(xlabel = None, ylabel = None, title = None, ylim = None):
    plt.gca().spines['left'].set_position('zero')
    plt.gca().spines['bottom'].set_position('zero')

    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    
    if ylim != None:
        plt.ylim(ylim)
```

# What is economics?

Economics is the study of choice under conditions of scarcity. Choice matters because choices involve **tradeoffs**. The first step to becoming an economist is deploying the word "tradeoff" as often as possible. 

We'll call the desirable things we're choosing between **goods** (lumping in services). As a child you might have weighed spending your pocket change on candy or on toys. In the common example of [guns and butter](https://en.wikipedia.org/wiki/Guns_versus_butter_model), economists say that a government, having only finite resources, must choose how much defense spending (guns) they might trade for more spending on civilian goods (butter), like social welfare programs. 

Sometimes a politician will insist that we can have our cake and eat it too. We can cut taxes and increase government revenue. We can provide universal (national\*) healthcare and reduce healthcare spending. Those are both theoretically possible. The why and how they could be possible must be put aside for now, but both claims ultimately have something to do with (in)**efficiency**. But once any government, company, household, or individaul runs efficiently, tradeoffs are still there. Economist's are fond of the phrase, "There ain't no such thing as a free lunch," memorably abbreviated as **TANSTAAFL**. You might get a free lunch, but did you have to wait in line? Did you have to listen to someone trying to sell [Tupperware](https://en.wikipedia.org/wiki/Tupperware#Tupperware_parties)? There was an **opportunity cost** because you could have done something else while you were waiting in line or you could have chosen different company or entertainment.  

Formally, the *opportunity cost* of something is the value of the best forgone alternative. 

# Production Possibility Frontiers

To wrap our minds around these ideas, we introduce the idea of a **production possibilities frontier**. This is the set of bundles of goods that can be produced when all resources are used and used efficiently. We plot one below.

```python
ppf = PPF(p1 = 1, p2 = 2, endowment = 10, good_names = ['Guns', 'Butter'])

ppf.plot()
plt.title("A linear PPF")
plt.gcf().set_size_inches(14,4.5)
```

##### The frontier is just the line. 

It shows what bundles of guns and butter are possible on the *frontier*. The PPF is just the blue line, not the enclosed triangular area. We could have ten guns and no butter, we could have five butter and no guns, or any combination in between. These would all be productively efficient, but in practice we can only choose one point. That's a tradeoff. Holding your five butter, you might want ten guns, but the cost is you'd have to give up those five butter. The slope is negative, signifying scarcity.

##### The slope is the opportunity cost of additional $x$ 
The slope of the PPF above is $-\frac{1}{2}$. The units of the slope are $\frac{\text{butter}}{\text{guns}}$. In plain English, that can be read as "butter per guns," Not so unlike prices you find in the store, like "dollars per apple." We can say  $\frac{1}{2}$ butter per gun is the gun price in terms of the only alternative good, making it an opportunity cost. 

##### Below the frontier is inefficient, beyond is unattainable

Returning to our idea of efficiency, the frontier is set of possible points given *productive effiency*. Anything below the frontier is inefficienct. Anything beyond the frontier is unattainable. 

```python
ppf.plot()
plt.title("Points Below, On, and Beyond the PPF")
plt.gcf().set_size_inches(14,4.5)

# efficient
point(6,2, color = 'C0', markersize = 24, annotate = '    Efficient')

# inefficient
point(4,2, color = 'gray')
point(3.3,1.1, annotate = 'Inefficient', markersize = 0)

# Unattainable
point(6,4, color = 'pink')
point(6.3,4, annotate = 'Unattainable', markersize = 0)
```

A point is inefficient if we can get more of either guns or butter without giving anything up. Above, the plotted inefficient point offers just as much butter as the plotted efficient point, but we can increase our level of guns by moving from to the efficient point. 


For an unusual PPF consider the following.

1.) You raise fish in one tank. The tank is sized for at most two fish. You can raise either piranhas or goldfish. However, if you mix golfish and piranhas, the piranha will eat the goldfish. What is the PPF? 

```python
point(0,2, color = 'C0')
point(2,0, color = 'C0')

plot_cleaner('Goldfish', 'Piranhas', 'Fish PPF <><')
```

Above, the PPF is just two points. There is no attainable point with golfish and piranhas because they don't get along (at no real fault of the goldfish). 

### The Robinson Crusoe Economy

**Example 2**. Alice and Bob split their time gathering and fishing. In an hour, Alice can gather two coconuts and Bob can gather just one. Alice can catch two fish. Bob doesn't know how to fish, so he can catch zero. What are the individual PPFs and what are the combined PPFs? Let's assume there are no diminishing returns--the PPFs are linear.

Alice's PPF is plotted below. 

```python
alice = PPF(p1 = 0.5, p2 = 0.5, good_names = ['Coconuts', 'Fish'])
alice.plot()
plt.title("Alice")
plt.show()
```

Bob's PPF is simple, but it's atypical--it's just the point 2 coconuts and 0 fish. 

```python
point(1,0, plt.gca(), color = 'C0')
plot_cleaner('Coconuts', 'Fish', 'Bob', ylim = (-.1,.5))
plt.show()
```

Alice can produce more coconuts and more fish than Bob given the same amount of time and resources. That means she has an **absolute advantage** in both goods. Still, as we'll later see, that doesn't mean the absolutely advantaged shouldn't specialize or that there will be no gains from trade. 

#### Trade

Let's think about trades that could make both better off. Suppose Bob only likes fish and suppose Alice only like coconuts. Bob can't fish so he has to trade his coconuts for fish produced by Alice.

What would make Bob better off? He'd have to accept the tiniest fraction of a fish for a coconut since he can't fish. But he'd of course prefer more fish per coconut, or a higher price for coconuts. 

What would make Alice better off? She'd prefer to give up as little fish per coconut. She'd prefer a lower price for coconuts. 

Imagine the price were 2 fish per coconut. 

What does that do for Bob's PPF?

```python
point(1,0, plt.gca(), color = 'C0')
plt.plot([0,1],[2,0], linestyle= 'dashed', color = 'black')
plot_cleaner('Coconuts', 'Fish', 'Bob')#, ylim = (-.1,))
plt.show()
```

What does it do for Alice's PPF, imagining she only produced fished and traded for coconuts?

```python
alice.plot()
point(0,2) # imagine she produces here and then trades along the dashed line
plt.plot([0,1],[2,0], linestyle= 'dashed', color = 'black')
plt.title("Alice")
plt.show()
```

This would make her worse off! The possibilities shrink! At a price of 2 fish per coconut, this is more expensive than her own opportunity cost. This is true for any price higher than 1 fish per coconut, which not coincidentally, is Alice's opportunity cost for producing coconuts. 

**Trading Prices:** The acceptable price range for a good is between the opportunity costs. 

For Alice and Bob, there are potential gains from trade for any price between 0 and 1 fish per coconut. 

#### Joint Production

How do we put the original PPFs together? Their **joint production possibilities frontier** is Bob's 1 Coconut + Whatever Alice produces. 

```python
plt.plot([1,3], [2,0])
plot_cleaner('Coconuts', 'Fish', title = 'Joint PPF')
plt.xlim(0,3.1)
plt.fill_between([0,1],2, color = (.92,.92,.92)) # fill in possibilities
plt.fill_between([1,3],[2,0], color = (.92,.92,.92))
plt.show()
```

Built into the joint PPF is the knowledge of how the two should specialize. It's taken for granted in the above that if we want 1 coconut, it is Bob who is doing that, so we don't sacrifice any of Alice's time (and therefore fish). 

##### Comparative Advantage and Specialization
In general, people should specialize according to their **comparative advantage**. Someone possesses a comparative advantage in a good when they have a lower opportunity cost of producing that good. Bob should gather coconuts because his opportunity cost is 0 fish. Gathering coconuts is costly for Alice, so if fish are desired, it should be Alice doing the fishing. This was the basis for the gains from trade above. 


If Bob and Alice work together, it's obvious how they should specialize. Bob will gather coconuts because that's all he's good for. Alice could spend her time gathering coconuts or fishing. Essentially, the joint PPF is Alice's PPF shifted by one additional coconut.  The exact mix they should choose depends on their preferences.

```python
plt.plot([1,3], [2,0])
alice.plot()
plot_cleaner('Coconuts', 'Fish', "Alice's PPF and the Joint PPF")
plt.show()
```

#### Bob Learns to Fish
Now suppose Bob learns to fish so that he can catch two fish in an hour. What is the joint PPF now?  

Firt, let's jut have a look at Bob's PPF. 

```python
bob = PPF(1,0.5, endowment = 1, good_names = ['Coconuts','Fish'])
bob.plot()
plt.title("Bob's New PPF")
```

### Constructing a joint PPF

Putting it together, we should see that now Bob's opportunity cost of fish is 0.5 coconuts. Compare that to Alice's opportunity cost of 1 coconut. Now Bob has the comparative advantage in fish, so he should specialize in fish and Alice should specialize in the coconuts. 

They specialize in their comparative advantage to the extent that that good is desired. 

Joint PPFs essentially connect a few extreme points. 

1.) Fish-intercept: how much fish could be produced if everyone fished? 

2.) Coconut-intercept: how many coconuts could be produced if everyone gathered coconuts? 

3.) Total specialization: How much of each good could be produced if everyone specialized in their comparative advantage? 

Connecting the points, we get points where one person is totally specialized, but another dabbles in both. Both produce according to their comparative advantage, but consumption preferences might dictate that we don't need so much fish (for example) as we'd have with total specialization.

```python
joint_ppf = JointPPF([alice, bob])

fig, ax = plt.subplots(2,2, figsize = (10,8))
joint_ppf.plot(ax[0,0])

#plt.fill_between([0,2],[4,2], 2, color = 'C1', alpha = 0.1)
#plt.fill_between([2,3],[2,0], color = 'C0', alpha = 0.1)
#plt.text(0.1,3, 'Alice', va = 'center')
#plt.text(2.1,1, 'Bob', va = 'center')

ax[0,0].annotate('Both fish', xy=(0, 4),  xycoords='data',
            xytext=(2.1, 3.5), #textcoords='axes fraction',
            arrowprops=dict(facecolor='blue'),
            size = 19)

joint_ppf.plot(ax[0,1])
ax[0,1].annotate('Both gather coconuts', xy=(3, 0),  xycoords='data',
            xytext=(1.1, 3.5), #textcoords='axes fraction',
            arrowprops=dict(facecolor='orange'),
            size = 19, ha = 'left')

joint_ppf.plot(ax[1,0])
ax[1,0].annotate('Bob fishes, Alice gathers\n(pure specialization)', xy=(2, 2),  xycoords='data',
            xytext=(1.1, 3.5), #textcoords='axes fraction',
            arrowprops=dict(facecolor='purple'),
            size = 19, ha = 'left')

joint_ppf.plot(ax[1,1])
ax[1,1].annotate('Alice gathers\nBob does a little of both', xy=(2.5, 1),  xycoords='data',
            xytext=(1.1, 3.4), #textcoords='axes fraction',
            arrowprops=dict(facecolor='green'),
            size = 19, ha = 'left')

plt.tight_layout()
plt.show()
            
```

The joint PPF essentially plots both individual PPFs, with each being shifted by the amount the other can produce according to their specialization. 

Below, the orange PPF is Alice's, shifted up by 2 fish (being how many fish Bob can produce if specializing) .

The blue PPF is Bob's, shifted right by 2 coconuts (being how many coconuts Alice can gather). 

The joint PPF outlines the frontier of these shifts. 

```python
joint_ppf.plot() #(ax[0,0])

plt.fill_between([0,2],[4,2], 2, color = 'C1', alpha = 0.1)
plt.fill_between([2,3],[2,0], color = 'C0', alpha = 0.1)
plt.text(0.04,2.3, r'Alice gathers coconuts $\rightarrow$', va = 'center')
plt.text(2.04,.21, r'Bob fishes$\uparrow$', va = 'center')

plt.show()
```

#### A three-person joint PPF. 

Let's introduce a third person, Conrad. Conrad has the following individual PPF.

His opportunity cost of a coconut is $\frac{3}{10}$ of a fish. 

```python
conrad = PPF(p1 = 0.1, p2 = 1/3, good_names = ['Coconuts', 'Fish'])
conrad.plot()
```

Now, Conrad has the lowest opportunity cost of a coconut ($\frac{3}{10}$). Alice has the next lowest at 1 fish per coconut. Bob has the highest cost at 2 fish per coconut.

Follow the frontier from the fish-intercept. The joint ppf encodes the following principle: The next coconut should be produced by the person who has the lowest opportunity cost. 

For coconuts 0-10, Conrad should produce coconuts. 

For coconuts 10-12, Alice should produce. Conrad's capabilities have been exhausted so we have to move on to Alice, our next producer. 

For coconuts 12-13, Bob should produce. Both Conrad and Alice have been exhausted, so we have to draft Bob. 

```python
tri_joint_ppf = JointPPF([alice, bob, conrad])

tri_joint_ppf.plot()
```
