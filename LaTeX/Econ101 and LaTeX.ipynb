# Writing a $\LaTeX$ Document from I-Python

This is a basic example showing how you can do everything from a Jupyter notebook. You can develop the examples in Python, save the figures, and write them to a tex file all in Python. 

```python
import os
import econ101 as econ
import matplotlib.pyplot as plt
```

```python
preamble = r"""\documentclass[12pt,notitlepage]{article}
\usepackage{101}
"""

main = r"""\begin{document}

\section*{Example Handout}
"""

end = r"""
\end{document}"""
```

# Matching Pennies

```python
matching_pennies = econ.Game((1,-1), (-1,1), (-1,1), (1,-1))
```

```python
matching_pennies.table(show_solution = False)

table0 = 'matching_p_no_sol.pdf'
plt.tight_layout()
plt.savefig(table0)
```

```python
matching_pennies.table(show_solution = True)
table1 = 'matching_p_with_sol.pdf'
plt.tight_layout()
plt.savefig(table1)
```

```python
mp_sec = r"""
\subsection*{Matching Pennies}

Player A's objective is to take the same action as B. B's objective is to take the opposite action.

\begin{center}
\includegraphics[width = .45\textwidth]{""" + table0 + r"""}\includegraphics[width = .45\textwidth]{""" + table1 + r"""}
\end{center}

There is no Nash equilibrium. 
"""
```

```python
# add mp_sec to main document piece
main += mp_sec
```

# Prisoner's Dilemma

```python
pris_dilemma = econ.Game((2,2), (-1,3), (3,-1), (0,0))

# Unsolved
pris_dilemma.table(show_solution = False)

table0 = 'pris_d_no_sol.pdf'
plt.tight_layout()
plt.savefig(table0)
plt.show()

# Solved game
pris_dilemma.table(show_solution = True)

table1 = 'pris_d_with_sol.pdf'
plt.tight_layout()
plt.savefig(table1)
plt.show()
```

```python
pd_sec = r"""
\subsection*{Prisoner's Dilemma}

Both A and B have a dominant strategy, action 1/confess. 

\begin{center}
\includegraphics[width = .45\textwidth]{""" + table0 + r"""}\includegraphics[width = .45\textwidth]{""" + table1 + r"""}
\end{center}

There is a unique Nash equilibrium. 
"""
```

```python
# add mp_sec to main document piece
main += pd_sec
```

# Save and Open Tex

```python
doc = preamble + main + end

# Save doc as tex file
filename = 'example101'
with open(filename + ".tex", 'w') as tf:
    tf.write(doc)
    
# Open tex file
os.system('open ' + filename + ".tex")
```
